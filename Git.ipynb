# **Task 3**

```python
# 1) Create new Git repo and commit main.txt on master
!git init multi_branch && cd multi_branch
!touch main.txt
!git add main.txt
!git commit -m "Add main.txt"

# 2) Create branches develop, f1, f2
!git branch develop
!git branch f1
!git branch f2

#3) Add develop.txt in develop branch and commit
!git checkout develop
!touch develop.txt
!git add develop.txt
!git commit -m "Add develop.txt"

# 4) Add f1.txt in f1 branch and commit
!git checkout f1
!touch f1.txt
!git add f1.txt
!git commit -m "Add f1.txt"

# 5) Add f2.txt in f2 branch and commit
!git checkout f2
!touch f2.txt
!git add f2.txt
!git commit -m "Add f2.txt"

# 6) Push all branches to GitHub
!git push origin master develop f1 f2

# 7) Delete f2 branch locally and remotely
!git branch -d f2
!git push origin --delete f2
```

# **Task 4**

```python
# 1) Add master.txt to master and commit
!touch master.txt
!git add master.txt
!git commit -m "Add master.txt"

# 2) Create branches public1, public2, private
!git branch public1
!git branch public2
!git branch private

# 3) Add public1.txt in public1 branch, commit, and merge into maste
!git checkout public1
!touch public1.txt
!git add public1.txt
!git commit -m "Add public1.txt"

!git checkout master
!git merge public1

# 4) Add public2.txt in public2 branch, commit, and merge into master
!git checkout public2
!touch public2.txt
!git add public2.txt
!git commit -m "Add public2.txt"

!git checkout master
!git merge public2

# 5) Edit master.txt in private branch and commit
!git checkout private
!echo "Edit from private branch" >> master.txt
!git add master.txt
!git commit -m "Edit master.txt in private branch"

# 6) Update public1 and public2 with new master code
!git checkout public1
!git merge master

!git checkout public2
!git merge master

# 7) Finally, sync private with updated master
!git checkout private
!git merge master
```

# **Task 5**

```python
# 1) Initialize Git repo with Git Flow
!git init gitflow_project && cd gitflow_project
!git flow init

# 2) Start a feature branch, add work, and finish it
!git flow feature start new-feature
!touch feature.txt
!git add feature.txt
!git commit -m "Add feature.txt"
!git flow feature finish new-feature

# 3) Start a release, finish it, and merge to master
!git flow release start v1.0
!git flow release finish v1.0

# 4) Create a hotfix branch for urgent changes and finish it
!git flow hotfix start urgent-fix
!touch urgent.txt
!git add urgent.txt
!git commit -m "Add urgent.txt via hotfix"
!git flow hotfix finish urgent-fix
```
